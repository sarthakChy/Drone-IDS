```python
import pandas as pd
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D


# --- Configuration ---
CSV_FILE_NAME = 'normal_corridor_base_data.csv'

# --- 1. Load Data ---
try:
    # Load the dataset from the provided CSV file path
    df = pd.read_csv(CSV_FILE_NAME)
    flight_id = df['flight_id'].iloc[0] if not df.empty else 'Unknown Flight'
    print(f"Successfully loaded {len(df)} data points for flight '{flight_id}' from {CSV_FILE_NAME}")

    # Define the columns required for plotting
    required_columns = ['pos_lon', 'pos_lat', 'pos_alt_rel', 'timestamp', 'att_roll', 'att_pitch']
    
    # Check if all required columns exist
    if not all(col in df.columns for col in required_columns):
        raise KeyError(f"One or more required columns are missing. Needed: {required_columns}")

    # Drop any rows with missing data in essential columns to prevent plotting errors
    df.dropna(subset=required_columns, inplace=True)
    
    if df.empty:
        print("Data is empty after removing rows with missing values. Cannot visualize.")
    else:
        # --- 2. 3D Flight Path Visualization ---
        print("Generating 3D Flight Path plot...")
        fig_3d = plt.figure(figsize=(12, 9))
        ax_3d = fig_3d.add_subplot(111, projection='3d')

        # Corrected column names based on your parser: 'pos_lon', 'pos_lat', 'pos_alt_rel'
        ax_3d.plot(df['pos_lon'], df['pos_lat'], df['pos_alt_rel'], label=f'Flight Path ({flight_id})')

        ax_3d.set_xlabel('Longitude', fontsize=10)
        ax_3d.set_ylabel('Latitude', fontsize=10)
        ax_3d.set_zlabel('Relative Altitude (m)', fontsize=10)
        ax_3d.set_title(f'3D Flight Path: {flight_id}', fontsize=14)
        ax_3d.legend()
        plt.show()

        # --- 3. Time-Series Telemetry Visualization ---
        print("Generating Time-Series Telemetry plot...")
        fig_ts, (ax1, ax2, ax3) = plt.subplots(3, 1, figsize=(16, 12), sharex=True)
        fig_ts.suptitle(f'Time-Series Telemetry: {flight_id}', fontsize=16)

        # Corrected column name for the x-axis: 'timestamp'
        # Plot 1: Altitude vs. Time
        ax1.plot(df['timestamp'], df['pos_alt_rel'], label='Altitude', color='blue')
        ax1.set_ylabel('Relative Altitude (m)')
        ax1.legend()
        ax1.grid(True)

        # Plot 2: Roll vs. Time
        ax2.plot(df['timestamp'], df['att_roll'], label='Roll', color='red')
        ax2.set_ylabel('Roll')
        ax2.legend()
        ax2.grid(True)

        # Plot 3: Pitch vs. Time
        ax3.plot(df['timestamp'], df['att_pitch'], label='Pitch', color='green')
        ax3.set_ylabel('Pitch')
        ax3.set_xlabel('Timestamp (ms from boot)')
        ax3.legend()
        ax3.grid(True)

        plt.tight_layout(rect=[0, 0.03, 1, 0.95]) # Adjust layout to make room for the suptitle
        plt.show()

except FileNotFoundError:
    print(f"Error: Could not find the file '{CSV_FILE_NAME}'")
    print("Please ensure the path is correct and the parser has been run.")
except KeyError as e:
    print(f"Error: A required column is missing from the CSV: {e}")
    print(f"Please check that your parser is outputting the correct headers.")
except Exception as e:
    print(f"An unexpected error occurred: {e}")
```
